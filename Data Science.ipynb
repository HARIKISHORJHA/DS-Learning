```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([2019,2020,2021],[10000,12000,15000])
plt.title("Practice Chart")
plt.ylabel("Income")
plt.xlabel("Year ")
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

plt.title("Practice Chart")
plt.ylabel("Users")
plt.xlabel("Year ")
year=[2019,2020,2021]
x_axis=np.arange(len(year))
Airtel_y=[10000,20000,18000]
Jio_y=[9000,15000,19000]
plt.plot(Airtel_y,label='Airtel')
plt.plot(Jio_y,label='Jio')
plt.xticks(x_axis, year)

plt.legend()
plt.show()
```
